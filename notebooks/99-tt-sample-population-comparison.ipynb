```python
from src.data import load_profiles
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = load_profiles()
```

```python
# filter non Berliners
data = data[data["district"] != " - "]
data = data[data["district"] != "Outsider"]
```

```python
# AGE
current = data["ageGroup"]
current = 100*(current.value_counts()/current.shape[0]).round(4).sort_index()
print(current)
current.plot(kind="bar")
```

```python
# GENDER
current = data["gender"]
current = 100*(current.value_counts()/current.shape[0]).round(4).sort_index()
print(current)
current.plot(kind="bar")
```

```python
# Mir steht folgendes zur Verfügung:

ve = "vehiclesOwned"
cols = []
vals = []
for col in data.columns:
    if ve in col:
        cols.append(col)
        vals.append(100*np.round(data[col].sum()/data[col].shape[0], 4))
plt.figure()
plt.bar(range(len(vals)),vals)
plt.xticks(range(len(vals)), cols, rotation=90)
plt.show()
print(vals)
```

```python
#Wie häufig nutzen Sie 
'''ratingLabels: [
      'Nie',
      'seltener als monatlich',
      '1-3 mal pro Monat',
      '1-3 mal pro Woche',
      '4-5 mal pro Woche',
      '(fast) täglich'''
# 3.4 Wege pro Tag

tr = "transportRatings"
for col in data.columns:
    if tr in col:
        plt.figure()
        plt.title(col)
        current = data[col]
        current = 100*(current.value_counts()/current.shape[0]).round(4).sort_index()
        print(current)
        current.plot(kind="bar")
```
