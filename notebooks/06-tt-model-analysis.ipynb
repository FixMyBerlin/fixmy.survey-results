# Hauptverkersstraßen

```python
from os.path import join
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams
from src.models.train_model import PROJECT_DIR
rcParams['figure.figsize'] = 14, 5
```

```python
df = pd.read_csv(join(PROJECT_DIR, "models", "MS_feature_importance.csv"),index_col=0)["0"].sort_values()
```

```python
fig = df.tail(20).plot(kind="barh", title="Größter positiver Einfluss")
```

```python
fig = df.head(20).plot(kind="barh", title="Größter negativer Einfluss")
```
